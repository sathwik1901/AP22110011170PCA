```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage import data
```

```python
img=data.camera()
```

```python
plt.imshow(img,cmap='gray')
plt.title("original")
plt.axis('off')
plt.show()
```

```python
X=img.reshape(-1,img.shape[1])
```

```python
def pca_recon(X, n):
    pca=PCA(n_components=n)
    X_pca=pca.fit_transform(X)
    X_inv=pca.inverse_transform(X_pca)
    return X_inv
```

```python
comps=[5,20,50,100]
imgs=[pca_recon(X,n) for n in comps]
```

```python
imgs=[img.reshape(img.shape) for img in imgs]
```

```python
fig,axs=plt.subplots(1,len(comps)+1,figsize=(15,5))
axs[0].imshow(img, cmap='gray')
axs[0].set_title("original")
axs[0].axis('off')
for i, n in enumerate(comps):
    axs[i+1].imshow(imgs[i], cmap='gray')
    axs[i+1].set_title(f"pca {n}")
    axs[i+1].axis('off')

plt.tight_layout()
plt.show()
```
